```python
url = "https://spitalstatistik.bagapps.ch/data/download/kzp21_KZ_TimeSerie.xlsx?v=1678279041"
sheet = 'KZ2008-KZ2021'
hospital_data = pd.read_excel(url, sheet_name=sheet)
```

```python
hospital_data.columns
```

```python
def search_and_replace(df, institution_name_dict):
    for search_string, replace_string in institution_name_dict.items():
      for index, row in df.iterrows():
        if search_string in row['Inst']:
            row['Inst']= replace_string
    return df
```

```python
hospital_data = search_and_replace(hospital_data, institution_name_dict)
```

```python
hospital_data.rename(columns={'Inst': 'hospital_name', 'JAHR':'year'}, inplace=True)
```

```python
hospital_data['hospital_name']
```

```python
concatenated_df = pd.read_csv('/content/drive/MyDrive/Module1_CDR/DF Merging/concatenated_df.csv', low_memory = False)
```

```python
merged_df = pd.merge(concatenated_df, hospital_data, on=['hospital_name', 'year'], how='outer')
```
