```python
#Import the data on the 2021 cases, cleaned by Lenja and in the shared drive under data
file_path = 'C:/Users/Gaëlle/Documents/_CAS applied data science/3. Module 3 Data analysis and machine learning/Project/unique names hospital.xlsx'
cases_name = pd.read_excel(file_path, sheet_name="Unique_names_cases_2020")
instit_name = pd.read_excel(file_path, sheet_name="Unique_names_instit_2020")
```

```python
cases_name
```

```python
import recordlinkage

indexer = recordlinkage.Index()
indexer.full()
candidates = indexer.index(cases_name, instit_name)
compare_cl = recordlinkage.Compare()
compare_cl.string('names_cases', 'name_instit', method='jarowinkler', threshold=0.85)
features = compare_cl.compute(candidates, cases_name, instit_name)
matches = features[features.sum(axis=1) > 0.85]
```

```python
matches.sum()
```

```python
correspondances = []
for idx1, idx2 in matches.index:
    nom_cases_name = cases_name.loc[idx1, 'names_cases']
    nom_instit_name = instit_name.loc[idx2, 'name_instit']
    correspondances.append((nom_cases_name, nom_instit_name))

for nom1, nom2 in correspondances:
    print(f"Nom dans df1 : {nom1}, Nom dans df2 : {nom2}")
```

```python
# Créez un DataFrame à partir de la liste de correspondances
df_correspondances = pd.DataFrame(correspondances, columns=['Nom_df1', 'Nom_df2'])

# Exportez le DataFrame dans un fichier CSV
df_correspondances.to_excel('correspondance_3.xlsx', index=False)
```
