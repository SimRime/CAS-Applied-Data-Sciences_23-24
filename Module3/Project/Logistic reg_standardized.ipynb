## Use dataset with the c-section threshold

```python
# Let's standardize all the variables
scaler = StandardScaler()
df_stand_array = scaler.fit_transform(df)
# back to a DataFrame
df = pd.DataFrame(df_stand_array, columns=df.columns)
```

## Creation of the X, Y, train and test

```python
# Creation of the dependant variable
target_field = 'threshold'
Y = df[target_field].copy()
```

```python
# Set of independant variables
X = df.drop(columns=['threshold','c_section_rate'])
```

```python
X.shape
```

```python
x = X.to_numpy().astype(np.float32)
y = Y.to_numpy().astype(np.float32)
```

```python
# train and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

### Fit the model

```python
logistic = linear_model.LogisticRegression(solver='sag', max_iter=200,
                             multi_class='ovr',)
logistic.fit(x_train, y_train)
```

```python
# Evaluate MSE, MAE, and R2 on train and test datasets
yy_test = logistic.predict(x_test)
yy_train = logistic.predict(x_train)

print('train R2 =', logistic.score(x_train, y_train))
print('test R2 =', logistic.score(x_test, y_test))
```

```python
# Coefficient matrix
cm=metrics.confusion_matrix( y_test, yy_test)
cm
```

```python
# Confusion Matrix Evaluation Metrics
from sklearn.metrics import classification_report
target_names = ['Below threshold', 'Above threshold']
print(classification_report(y_test, yy_test, target_names=target_names))
```
