## Function to Import URL Links

```python
def import_url(url_lst,year_tuple)    :
    '''Function to import multiple URL in a single dataframe and add a column
    corresponding to the year
    parameters:
        url_lst = list of url adresses
        year_tuple = tuple of the year to add in the column '''

    i=0
    df_final = pd.DataFrame()

    for year in range(year_tuple[0],year_tuple[1]):
        # Remplacez 'lien_annee' par le lien spécifique pour chaque année
        url = url_lst[i]

        # Utilisez pd.read_html() pour importer la table à partir du lien
        tables = pd.read_html(url, encoding='utf-8')

        # Sélectionnez la table appropriée (peut nécessiter un examen préalable)
        df = tables[0]

        # Ajoutez une colonne 'year' avec l'année correspondante
        df['year'] = year

        # Ajoutez le DataFrame à la liste
        df_final = pd.concat([df_final, df], ignore_index=True)
        i+=1

    return df_final
```

Use of the function "import_url" for each data

```python
df_cesarienne = import_url(url_c_section, year_tuple)
kidney = import_url(url_kidney, year_tuple)
accouchement = import_url(url_delivery, year_tuple)
```

Cleaning and merging of the Dataframes

```python
#Rename the columns of the dataframes
kidney.rename(columns={'hôpital': 'hospital', 'nombre de cas':'cases_kidney'}, inplace=True)
df_cesarienne.rename(columns={'hôpital': 'hospital', 'nombre de cas':'cases_c-section'}, inplace=True)
accouchement.rename(columns={'hôpital': 'hospital', 'nombre de cas':'delivery'}, inplace=True)

#Merger les df en fonction de year, hospital et canton
df_global = accouchement.merge(df_cesarienne, on=['year', 'hospital','canton'], how='left')
df_global = df_global.merge(kidney, on=['year', 'hospital','canton'], how='left')
df_global = df_global[df_global['canton'] != 'canton']
```

Add a column if the Canton is Swiss german or Latin and the same for university

```python
#Définition des cantons Latins
Canton_RO = ['GE','VD','NE','JU','FR','VS','TI']

#Définition des hopitaux universitaires
nom_hopitaux = ['Les Hôpitaux Universitaires de Genève HUG',
                'CHUV Centre Hospitalier Universitaire Vaudois',
                'Inselspital Bern','Insel Gruppe AG (universitär)',
                'Universitätsspital Zürich',
                'Universitätsspital Basel',
               'UniversitÃ¤tsspital ZÃ¼rich',
               'CHUV Centre Hospitalier Universitaire Vaudois',
               'UniversitÃ¤tsspital Basel',
                'Les HÃ´pitaux Universitaires de GenÃ¨ve HUG',
                'Insel Gruppe AG (universitÃ¤r)']

#Ajout de variable pour définir si canton latin/allémanique et hôpital universitaire
df_global['Reg'] = df_global['canton'].apply(lambda x: 'RO' if x in Canton_RO else 'AL')
df_global['Uni'] = df_global['hospital'].apply(lambda x: 1 if x in nom_hopitaux else 0)
```

Cleaning up numbers with a space after thousands and adding a column with the percentage of c-section

```python
#Conversion des valeurs string en int
df_global['delivery'] = df_global['delivery'].apply(lambda x: x.replace('\xa0', ''))
df_global['delivery'] = pd.to_numeric(df_global['delivery'], errors='coerce')

df_global['cases_c-section'] = df_global['cases_c-section'].apply(lambda x: x.replace('\xa0', '') if isinstance(x, str) else x)
df_global['cases_c-section'] = pd.to_numeric(df_global['cases_c-section'], errors='coerce')

df_global['cases_kidney'] = df_global['cases_kidney'].apply(lambda x: x.replace('\xa0', '') if isinstance(x, str) else x)
df_global['cases_kidney'] = pd.to_numeric(df_global['cases_kidney'], errors='coerce')

#Ajout d'une ligne pour le pourcenatage de C-Section
df_global['percentage_c-section'] = df_global['cases_c-section']/df_global['delivery']
```

# Importation of the hopsitals Data

```python
url = "https://spitalstatistik.bagapps.ch/data/download/kzp21_KZ_TimeSerie.xlsx?v=1678279041"
sheet = 'KZ2008-KZ2021'
df_hopitaux = pd.read_excel(url, sheet_name=sheet)
```

Deletion of data prior to 2015

```python
df_hopitaux = df_hopitaux[df_hopitaux['JAHR']>2014]
```

Rename columns and process column with private percentage

```python
columns = ['JAHR','KT','Inst','PersA','PersAWB','PersP','PersMT','PersT','Ops','Gebs','pPatLKP']

df_hopitaux = df_hopitaux[columns]

df_hopitaux['Inst'] = df_hopitaux['Inst'].replace('Inselspital Bern', 'Insel Gruppe AG (universitär)')

columns = ['JAHR','KT','Inst','Doctor','Doctor_in_formation','Nurse','Other_Nurse','Total_staff','Operation_rooms','Delivery_rooms','Private_division']

df_hopitaux.columns = columns

df_hopitaux.rename(columns={'Inst': 'hospital', 'JAHR':'year', 'KT':'canton'}, inplace=True)

# Use a lambda function to conditionally divide values by 100 if greater than one
# (The calculation has changed between 2015 and 2021)
df_hopitaux['Private_division'] = df_hopitaux['Private_division'].apply(lambda x: x / 100 if x > 1 else x)
```

# Merging of the two dataframes and exportation

```python
df_global = df_global.merge(df_hopitaux, on=['year', 'hospital', 'canton'], how='left')

# Save the DataFrame to a CSV file
df_global.to_csv('hospital_cases.csv', index=False)  # Use index=False to omit writing row numbers as index
```
